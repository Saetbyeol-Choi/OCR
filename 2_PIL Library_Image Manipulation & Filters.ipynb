```python
import PIL
from PIL import Image
from IPython.display import display
file="monalisa.gif"
image=Image.open(file)
```

```python
help(image.copy)
```

```python
help(image.save)
```

```python
# The file was originally a GifImageFile, but if we save it with a .png format and read it in again we'll get a different kind of file
image.save("monalisa.png")
image=Image.open("monalisa.png")
import inspect
inspect.getmro(type(image))
```

```python
from PIL import ImageFilter
help(ImageFilter)
```

```python
# Try and apply the BLUR filter.
# Before we do this we have to convert the image to RGB mode.
# Images like gifs are limited in how many colors can be displayed at once based on the size of the pallet.
image=image.convert('RGB') # this stands for red, green blude mode
blurred_image=image.filter(PIL.ImageFilter.BLUR)
display(blurred_image)
```

```python
print("{}x{}".format(image.width, image.height))
```

```python
help(image.crop)
```

```python
# start witht the left at, say 50 pixels, and the top at 0 pixels, then walk to the right another 190 pixels, and set the lower bound to say 150 pixels
display(image.crop((50,0,190,150)))
```

```python
# crop() only returns a copy of the image, and doesn't change the image itself.
# Draw the bouding box using the ImageDraw object.
from PIL import ImageDraw
drawing_object=ImageDraw.Draw(image)
drawing_object.rectangle((50,0,190,150), fill = None, outline ='red')
display(image)
```
